```python
import pandas as pd
```

```python
file = "purchase_data.csv"
purchase_data_df = pd.read_csv(file)
purchase_data_df.head()
```

```python
count = len(purchase_data_df["SN"].unique())
count1 = [{"Total Players" : count}]
noofplayers_df = pd.DataFrame(count1)
noofplayers_df
```

```python
uniitems = len(purchase_data_df["Item ID"].unique())
uniitems
avgprice = purchase_data_df["Price"].mean(axis=0)
avgprice
noofpurchases = purchase_data_df["Purchase ID"].count()
noofpurchases
totalrevenue = purchase_data_df["Price"].sum()
totalrevenue
summary_df = [{"Number of Unique Items": uniitems, "Average Price": avgprice, "Number of Purchases" : noofpurchases, "Total Revenue" : totalrevenue}]
summary_df = pd.DataFrame(summary_df)
#summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
#summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,}".format)
#summary_df = summary_df.style.set_properties(**{'text-align':'left'})
summary_df
```

```python
maleplayer = purchase_data_df.loc[purchase_data_df["Gender"]=="Male"]
femaleplayer = purchase_data_df.loc[purchase_data_df["Gender"]=="Female"]
otherplayer = purchase_data_df.loc[purchase_data_df["Gender"]=="Other / Non-Disclosed"]
malecount = len(maleplayer["SN"].unique())
femalecount = len(femaleplayer["SN"].unique())
othercount = len(otherplayer["SN"].unique())
percentagemale = (malecount/count)*100
percentagefemale = (femalecount/count)*100
percentageother = (othercount/count)*100

summary1_df = pd.DataFrame([{"Gender":"Male", "Total Count": malecount, "Percentage of Players": percentagemale}, {"Gender":"Female", "Total Count": femalecount, "Percentage of Players": percentagefemale}, {"Gender":"Other / Non-Disclosed", "Total Count": othercount, "Percentage of Players": percentageother}])
summary1_df = summary1_df.set_index("Gender")
summary1_df.index.name=None
summary1_df
```

```python
groupedpurchase = purchase_data_df.groupby('Gender').agg({"Purchase ID": "count", "Price": ["mean", "sum"], "SN": "nunique"})
groupedpurchase.columns=["Purchase Count", "Average Purchase Price", "Total Purchase Value", "SN"]
groupedpurchase["Avg Total Purchase per Person"] = groupedpurchase["Total Purchase Value"]/groupedpurchase["SN"]
groupedpurchase.drop(columns=["SN"])
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
group_labels = ["<10","10-14","15-19","20-24","25-29", "30-34", "35-39", "40+"]
pd.cut(purchase_data_df["Age"], bins, labels=group_labels).head()
purchase_data_df["Age Range"] = pd.cut(purchase_data_df["Age"], bins, labels=group_labels, include_lowest=True)
purchase_data_df
Agerange = purchase_data_df.groupby("Age Range").agg({"SN": "nunique"})
Agerange.columns=["Total Count"]
Agerange["Percentage of Players"] = Agerange["Total Count"]/Agerange["Total Count"].sum()
Agerange.head(10)
```

```python
Agerange1 = purchase_data_df.groupby("Age Range").agg({"Purchase ID": "count", "Price": ["mean", "sum"], "SN": "nunique"})
Agerange1.columns=["Purchase Count", "Average Purchase Price", "Total Purchase Value", "SN"]
Agerange1["Avg Total Purchase per Person"] = Agerange1["Total Purchase Value"]/Agerange1["SN"]
Agerange1.drop(columns=["SN"])
Agerange1.head(10)
```

```python
Agerange1 = purchase_data_df.groupby("SN").agg({"Purchase ID": "count", "Price": ["mean", "sum"]})
Agerange1.columns=["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
Agerange1 = Agerange1.sort_values("Total Purchase Value", ascending=False)
Agerange1.head(10)
```

```python
Agerange2 = purchase_data_df.groupby(["Item ID", "Item Name"]).agg({"Purchase ID": "count", "Price": ["mean", "sum"]})
Agerange2.columns=["Purchase Count", "Item Price", "Total Purchase Value"]
Agerange2 = Agerange2.sort_values("Purchase Count", ascending=False)
Agerange2.head(10)
```

```python
Agerange3 = purchase_data_df.groupby(["Item ID", "Item Name"]).agg({"Purchase ID": "count", "Price": ["mean", "sum"]})
Agerange3.columns=["Purchase Count", "Item Price", "Total Purchase Value"]
Agerange3 = Agerange2.sort_values("Total Purchase Value", ascending=False)
Agerange3.head(10)
```
